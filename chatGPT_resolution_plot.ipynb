```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset

# Example Data
I = np.random.rand(50, 50)  # 50x50 low-res data
epsilons = np.linspace(0, 1, I.shape[1])
VBs = np.linspace(0, 1, I.shape[0])

# Create Figure
fig, ax = plt.subplots(figsize=(8, 6))

# Main imshow plot
img = ax.imshow(I, extent=[epsilons[0], epsilons[-1], VBs[0], VBs[-1]], aspect='auto', origin='lower')
fig.colorbar(img, ax=ax)

ax.set_title("Main Plot with High-Resolution Zoom")
ax.set_xlabel('epsilon')
ax.set_ylabel('V')

# Define zoom area (e.g., focus on a small square region)
eps_min, eps_max = 0.4, 0.6  # Epsilon range
v_min, v_max = 0.4, 0.6  # V range

# Create a high-res version of the zoomed region
zoom_size = 100  # More resolution
x_zoom = np.linspace(eps_min, eps_max, zoom_size)
y_zoom = np.linspace(v_min, v_max, zoom_size)
I_zoom = np.random.rand(zoom_size, zoom_size)  # Simulated high-res data

# Inset plot (zoomed-in square)
axins = inset_axes(ax, width="30%", height="30%", loc="upper right")  # Size of the inset

# High-res zoomed-in plot
img_zoom = axins.imshow(I_zoom, extent=[eps_min, eps_max, v_min, v_max], aspect='auto', origin='lower')

axins.set_xticks([])  # Remove ticks for clarity
axins.set_yticks([])

# Mark the zoomed region with a rectangle
ax.plot([eps_min, eps_max, eps_max, eps_min, eps_min], [v_min, v_min, v_max, v_max, v_min], 'r-', linewidth=2)

# Connect main plot and inset with lines
mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="red")

plt.show()
```
